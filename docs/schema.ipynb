### **Real Estate Transactions Schema**

| Field Name              | Description                                                   | Data Type   |
|-------------------------|---------------------------------------------------------------|-------------|
| Transaction Number      | Unique identifier for the transaction                         | String      |
| Transaction Date        | Date and time of the transaction                              | Date & Time |
| Property ID             | Unique identifier for the property                            | Integer     |
| Transaction Type        | Type of transaction (Sales, Gifts, Mortgage)                 | String      |
| Transaction Sub Type    | Subtype of the transaction (e.g., Delayed Sell)               | String      |
| Registration Type       | Type of registration (Ready, Off-Plan)                        | String      |
| Is Free Hold?           | Indicates if the property is freehold                         | String      |
| Usage                   | Usage of the property (Commercial, Residential)              | String      |
| Area                    | Area of Dubai where the property is located                  | String      |
| Property Type           | General type of property (Unit, Land, Building)              | String      |
| Property Sub Type       | Specific subtype of property (e.g., Flat, Villa)             | String      |
| Amount                  | Transaction amount in AED                                     | Float       |
| Transaction Size (sq.m) | Area of the transaction in square meters                      | Float       |
| Property Size (sq.m)    | Area of the property in square meters                         | Float       |
| Room(s)                 | Layout of the property (e.g., Studio, 1 B/R)                 | String      |
| Parking                 | Number of parking spaces                                      | Integer     |
| Nearest Metro           | Name of the nearest metro                                     | String      |
| Nearest Mall            | Name of the nearest mall                                      | String      |
| Nearest Landmark        | Name of the nearest landmark                                  | String      |
| No. of Buyer            | Number of buyers involved in the transaction                 | Integer     |
| No. of Seller           | Number of sellers involved in the transaction                | Integer     |
| Master Project          | Name of the master project                                   | String      |
| Project                 | Name of the project                                          | String      |

### **Brief Overview**

- The **Transactions Dataset** provides granular details of property sales and related activities, capturing financial, spatial, and property-specific information.
- The **Rental Dataset** (Ejari data) focuses on leasing agreements, giving insights into rental trends and property usage over time.
- These datasets are complemented by macroeconomic indicators, such as GDP, population, and tourism statistics, to analyze their impact on real estate trends.
